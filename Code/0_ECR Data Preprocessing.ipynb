### Import necessary packages and read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyxlsb import open_workbook as open_xlsb

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 500)
pd.options.mode.chained_assignment = None
```

```python
# Define a function to read xlsb file into a pandas DataFrame
def read_xlsb(file):
    data = []
    with open_xlsb(file) as wb:
        with wb.get_sheet(1) as sheet:
            for row in sheet.rows():
                data.append([item.v for item in row])  # extracting value from the cell item

    df = pd.DataFrame(data[1:], columns=data[0])
    
    return df

# Read the xlsb file into a pandas DataFrame
df = read_xlsb("../Data/Original/1994-2022_ECA_Data.xlsb")

# Drop the columns where all elements are None
df = df.dropna(how='all', axis=1)

# Change opdate column to a proper data
df['Opdate'] = pd.to_datetime(df['Opdate'], unit='D', origin='1898-12-30')

# Get basic information about the DataFrame
df.info()
```

```python
# Columns to exclude from rounding and conversion
exclude_columns = [
    "Patient_Number", "Sex", "Age", "Year", "Operation_Date", "IC_cm",
    "lesion_text", "Lesion_Coding", "other_text_neck", "other_text_media",
    "other_text_abdo", "pTNM7_1", "pStage7_1", "pT7", "pN7", "pM7", "pStage7"
]

# Identifying the columns to modify
columns_to_modify = [col for col in df.columns if col not in exclude_columns]

# Identifying the numerical columns to modify
numerical_columns_to_modify = [col for col in columns_to_modify if df[col].dtype in ['float64', 'int64']]

# Rounding the values to the nearest whole number and converting to integers
for col in numerical_columns_to_modify:
    df[col] = df[col].apply(lambda x: round(x) if pd.notnull(x) else x)

# Defining the primary sites based on the "Lesion_Coding" column
primary_sites_mapping = {1: 'upper', 2: 'mid', 3: 'lower'}
df['Primary_Site'] = df['Lesion_Coding'].map(primary_sites_mapping)
```
